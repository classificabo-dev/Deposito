```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar o dataset Gapminder do Seaborn
#df = sns.load_dataset("Gapminder")
df = pd.read_csv("Gapminder.csv")

# Filtrar o último ano disponível
df_ultimo = df[df["year"] == df["year"].max()]


```

```python
# Calcular a mediana da expectativa de vida por continente
medianas = df_ultimo.groupby("country", as_index=False)["lifeExp"].median()

# Plot do gráfico de bolhas
plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=df_ultimo,
    x="gdpPercap",
    y="lifeExp",
    size="pop",
    hue="continent",
    sizes=(40, 1200),
    alpha=0.6,
    edgecolor="black"
)

# Adicionar a linha de mediana (uma por continente)
sns.lineplot(
    data=medianas,
    x="continent",
    y="lifeExp",
    color="red",
    linewidth=2.5,
    marker="o",
    label="Mediana por Continente"
)
```

```python
# Ajustar o eixo X (escala logarítmica para PIB)
plt.xscale("log")

# Títulos e rótulos
plt.title("PIB per Capita vs Expectativa de Vida — com Linha de Mediana por Continente", fontsize=14)
plt.xlabel("PIB per Capita (escala logarítmica)")
plt.ylabel("Expectativa de Vida")
plt.legend(title="continent", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
